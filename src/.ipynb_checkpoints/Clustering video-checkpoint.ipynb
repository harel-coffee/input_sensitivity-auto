```python
video_meta = pd.read_csv('../data/videolist.csv', delimiter =',')
video_meta.head()
```

```python
video_meta['log_bit_rate'] = np.log(video_meta['bit_rate'])
video_meta['log_size'] = np.log(video_meta['size'])
video_meta['log_width'] = np.log(video_meta['width'])
video_meta['log_height'] = np.log(video_meta['height'])
video_meta['log_duration'] = np.log(video_meta['duration'])
```

```python
video_meta['abbrsize'].values
```

```python
data = np.array(video_meta[['log_size', 'duration', 'log_width', 'log_bit_rate',
       'log_height', 'r_frame_rate', 'abbrsize_num', 'dynamic', 'camera_move',
       'outdoor', 'zoom', 'stability_score']], dtype=float)
data[0:5]
```

```python
d = StandardScaler(data, with_std=True, with_mean=True)
d.copy
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(d.copy)
```

```python
kmeans.labels_
```

```python
video_meta['metadata_group'] = kmeans.labels_
video_meta.to_csv('test.csv')
video_meta
```

```python
col = ['blue','red','green','orange']

plt.figure(figsize=(15,15))
plt.scatter(video_meta['width'],
            video_meta['height'], 
            color = [col[k] for k in video_meta['metadata_group']])
plt.show()
```

### Simple example with sklearn

```python
X = np.array([[5,3],
     [10,15],
     [15,12],
     [24,10],
     [30,45],
     [85,70],
     [71,80],
     [60,78],
     [55,52],
     [80,91],])
```

```python
plt.scatter(X[:,0],X[:,1], label='True Position')
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
print(kmeans.cluster_centers_)
```

```python
print(kmeans.labels_)
```

```python
plt.scatter(X[:,0],X[:,1], c=kmeans.labels_, cmap='rainbow')
```
