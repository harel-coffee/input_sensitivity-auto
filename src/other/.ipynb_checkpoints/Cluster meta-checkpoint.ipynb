Import all metrics

```python
meta = pd.read_csv("../../data/ugc_meta/all_features.csv").set_index('FILENAME')
#mos = pd.read_csv('../ugc_meta/mos.csv', delimiter =';').set_index('FILENAME')
#del mos['category']
#meta = meta.set_index('FILENAME')
#meta = meta.join(mos, how='outer')#.to_csv("../ugc_meta/test.csv")
meta['category']=[str(meta.index[i]).split('_')[0] for i in range(meta.shape[0])]
del meta['NOISE_DMOS']
#meta['NOISE_DMOS']=meta['NOISE_DMOS'].replace('[NULL]','0').replace('nan','0')
#meta['NOISE_DMOS']=meta['NOISE_DMOS'].replace('0',str(np.mean(np.array(meta['NOISE_DMOS'],float))))
meta
meta = meta.fillna(0)
cat_tab = pd.Series(meta['category'].values).unique()
meta['cat_num'] = [np.where(cat_tab==meta['category'][i])[0][0] for i in range(len(meta['category']))]
del meta['category']
for col in meta.columns:
    inter = np.array(meta[col],float)
    meta[col] = (inter-np.mean(inter))/np.std(inter)
meta
```

```python
std_meta = np.array(meta, float)
```

# KMeans

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(std_meta)
meta['clusters'] = kmeans.labels_
meta['category']=[str(meta.index[i]).split('_')[0] for i in range(meta.shape[0])]
meta['quality']=[str(meta.index[i]).split('_')[1].split('-')[0] for i in range(meta.shape[0])]
meta
```

```python
pd.crosstab(meta['clusters'], meta['quality'])
```

```python
pd.crosstab(meta['clusters'], meta['category'])
```

```python
col = ['gray','red','green','black']

plt.figure(figsize=(15,15))
plt.scatter(meta['WIDTH'],
            meta['HEIGHT'], 
            color = [col[k] for k in meta['clusters']])
plt.show()
```

# PCA 

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2, svd_solver='full')
tab = pca.fit_transform(std_meta)
x = [tab[i][0] for i in range(len(tab))]
y = [tab[i][1] for i in range(len(tab))]

col = ['gray','red','green','black']

plt.figure(figsize=(16,8))
plt.title("Axis (PCA) Color Kmeans classes")
plt.scatter(x, y, color = [col[k] for k in meta['clusters']])
lab=np.array(meta.index,str)
#for i in range(len(x)):
#    plt.text(x[i], y[i], lab[i])
plt.show()
```
