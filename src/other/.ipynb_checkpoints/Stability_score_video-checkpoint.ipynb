## Stability scores

```python
def diff_frames(frame1, frame2, nb_bins=255):
    
    red_fr1 = np.histogram(frame1[:,:,0].reshape(-1,1), bins=nb_bins)
    red_fr2 = np.histogram(frame2[:,:,0].reshape(-1,1), bins=nb_bins)
    
    green_fr1 = np.histogram(frame1[:,:,1].reshape(-1,1), bins=nb_bins)
    green_fr2 = np.histogram(frame2[:,:,1].reshape(-1,1), bins=nb_bins)
    
    blue_fr1 = np.histogram(frame1[:,:,2].reshape(-1,1), bins=nb_bins)
    blue_fr2 = np.histogram(frame2[:,:,2].reshape(-1,1), bins=nb_bins)
    
    return np.mean(np.abs(red_fr1[0]-red_fr2[0])
                  +np.abs(green_fr1[0]-green_fr2[0])
                  +np.abs(blue_fr1[0]-blue_fr2[0]))/np.mean(red_fr1[0]+green_fr1[0]+blue_fr1[0])

def get_diff_score(name_video, delta = 1):

    diff_scores = []
    
    vid = cv2.VideoCapture(name_video)

    prev_frame = None
    frame = None
    
    ind = 0
    
    while(vid.isOpened()):
        
        compare_frame = (ind % delta == 0)
        
        ret, frame = vid.read()

        if ret:

            #cv2.imshow('Video', frame)

            if type(prev_frame)!=type(None) and ind %delta ==0 :
                diff_scores.append(diff_frames(prev_frame,frame))

            #cv2.waitKey(50)

        else:
            break

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
        
        if compare_frame:
            prev_frame = frame

        ind+=1

    vid.release()
    cv2.destroyAllWindows()
    
    return diff_scores
```

```python
sintel = cv2.VideoCapture('../sintel.mp4')

address = "../videos/sintel/"

cut = [100, 200, 275, 375, 425, 550, 675, 850, 1000]
compteur = 0

fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter(address+'sintel0.mp4',fourcc, 20.0, (int(sintel.get(3)), int(sintel.get(4))))

index_cut = 0

while(sintel.isOpened()):
    
    compteur+=1
    
    ret, frame = sintel.read()

    if ret:
        
        out.write(frame)
        
        cv2.imshow('Sintel - video 4', frame)
        
        if compteur==cut[min(index_cut,len(cut)-1)]:
            index_cut+=1
            out.release()
            out = cv2.VideoWriter(address+'sintel'+str(index_cut)+'.mp4',
                                  fourcc, 20.0, (int(sintel.get(3)), int(sintel.get(4))))
            
    else:
        break
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

out.release()
sintel.release()
cv2.destroyAllWindows()
```
