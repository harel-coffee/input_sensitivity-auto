```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

# Comparison with gaming video

```python
def elapsedtime_to_sec(el):
    tab = el.split(":")
    return float(tab[0])*60+float(tab[1])

def pow10(x):
    return np.round(np.power(10,float(x)),2)

title_size=20
label_size = 15
default = ['ultrafast','superfast','veryfast','faster','fast','medium',
           'slow', 'slower','veryslow','placebo']

col = ['red','orange','coral','lightgreen', 'green', 'blue', 
       'lightblue', 'cyan','purple', 'pink', 'darkgrey']

x_text = [*map(pow10, [-1.75+k*0.25 for k in range(10)])]
y_text = np.array(np.array([*map(pow10, [4.8+k*0.1 for k in range(10)])],dtype=int)/1000,int)

fig = plt.figure(figsize=(20,9))
        
data = pd.read_table("../../data/gaming_old_result.csv",
             delimiter=',')
data['etime'] = [*map(elapsedtime_to_sec, data['elapsedtime'])]
#data['el_sec'] = [*map(elapsedtime_to_sec, data['elapsedtime'])]

ax = fig.add_subplot(1,2,1)
plt.xlim((-0.5,1.7))
plt.ylim((6.7,7.16))
plt.title("Old configurations (n=1152)", size=title_size)
#new_text = [*map(pow10, [-0.5,0,0.5,1,1.5])]
ax.set_xticklabels(x_text)
ax.set_yticklabels(y_text)

plt.xlabel('Elapsed time (seconds, log scale)',size=label_size)
plt.ylabel('Size (MB, log scale)',size=label_size)
plt.scatter(np.log10(data.query('configurationID<=1152')['etime']), 
            np.log10(data.query('configurationID<=1152')['size']),color="black")

for i in range(len(default)):
    ax.scatter(np.log10(data.query('configurationID=='+str(i+1+1152))['etime']), 
               np.log10(data.query('configurationID=='+str(i+1+1152))['size']), 
               s= 100, marker ='*',color= col[i])
    ax.text(np.log10(data.query('configurationID=='+str(i+1+1152))['etime']),
            np.log10(data.query('configurationID=='+str(i+1+1152))['size']),
            default[i], color=col[i], size = 10)


old_range_time = np.percentile(data['etime'], 95)-np.percentile(data['etime'], 5)
old_range_size = np.percentile(data['size'], 95)-np.percentile(data['size'], 5)

data = pd.read_table("../../data/gaming_new_result.csv", delimiter=',')

data['etime'] = [*map(elapsedtime_to_sec, data['elapsedtime'])]

for i in range(len(default)):
    ax.scatter(np.log10(data.query('configurationID=='+str(i+1))['etime']),
               np.log10(data.query('configurationID=='+str(i+1))['size']), 
               s= 100, marker ='*',color= col[i])
    ax.text(np.log10(data.query('configurationID=='+str(i+1))['etime']),
            np.log10(data.query('configurationID=='+str(i+1))['size']),
            default[i], color=col[i], size = 15)
    
ax = fig.add_subplot(1,2,2)

#data = pd.read_table("../group_paper/gaming_new_result.csv", delimiter=',')

#data['etime'] = [*map(elapsedtime_to_sec, data['elapsedtime'])]

plt.title("New configurations (n=200)",size=title_size)

plt.xlabel('Elapsed time (seconds, log scale)',size=label_size)
plt.ylabel('Size (MB, log scale)',size=label_size)
plt.scatter(np.log10(data['etime']), np.log10(data['size']),color="black")

for i in range(len(default)):
    ax.scatter(np.log10(data.query('configurationID=='+str(i+1))['etime']),
               np.log10(data.query('configurationID=='+str(i+1))['size']), 
               s= 100, marker ='*',color= col[i])
    ax.text(np.log10(data.query('configurationID=='+str(i+1))['etime']),
            np.log10(data.query('configurationID=='+str(i+1))['size']),
            default[i], color=col[i], size = 15)


ax.set_xticklabels(x_text)
ax.set_yticklabels(y_text)

plt.xlim((-0.5,1.7))
plt.ylim((6.7,7.16))
      
plt.savefig("../../results/config_comparison.png")
plt.show()

new_range_time = np.percentile(data['etime'], 95)-np.percentile(data['etime'], 5)
new_range_size = np.percentile(data['size'], 95)-np.percentile(data['size'], 5)

print(old_range_time)
print(new_range_time)
print(old_range_size)
print(new_range_size)
print("ratio time : "+str(new_range_time/old_range_time))
print("ratio size : "+str(new_range_size/old_range_size))
```

# Default configuration

```python
def elapsedtime_to_sec(el):
    tab = el.split(":")
    return float(tab[0])*60+float(tab[1])

def pow10(x):
    return np.round(np.power(10,float(x)),2)

vids = os.listdir("../../data/old/")
v_names = [vid for vid in vids if vid[len(vid)-4:len(vid)] not in ['.csv', '.ods', '.pdf','ntel'] ]

default = ['ultrafast','superfast','veryfast','faster','fast','medium',
           'slow', 'slower','veryslow','placebo']

col = ['red','orange','coral','lightgreen', 'green', 'blue', 
       'lightblue', 'cyan','purple', 'pink', 'darkgrey']

fig = plt.figure(figsize=(20,20))

for l in range(5):
    for j in range(5):
        
        v = v_names[l*5+j]
        
        data = pd.read_table("../../data/old/"+v+"/result_mean.csv",
                     delimiter=',')

        #data['el_sec'] = [*map(elapsedtime_to_sec, data['elapsedtime'])]

        ax = fig.add_subplot(5,5,l*5+j+1)

        plt.title(v)
        ax.set_xticks([-0.5,0,0.5,1,1.5])

        #plt.xlabel('Elapsed time (log scale)')
        #plt.ylabel('Size')
        plt.scatter(np.log10(data.query('configurationID<=1152')['etime']), 
                    data.query('configurationID<=1152')['size'],
                   color="black")

        for i in range(len(default)):
            ax.scatter(np.log10(data.query('configurationID=='+str(i+1+1152))['etime']),
                       data.query('configurationID=='+str(i+1+1152))['size'], 
                       s= 100, marker ='*',color= col[i])
            ax.text(np.log10(data.query('configurationID=='+str(i+1+1152))['etime']),
                data.query('configurationID=='+str(i+1+1152))['size'],
                default[i], color=col[i],
                 size = 10)

        new_text = [*map(pow10, [-0.5,0,0.5,1,1.5])]
        ax.set_xticklabels(new_text)
plt.show()
```
