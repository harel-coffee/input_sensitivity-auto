```python
import pandas as pd
```

```python
meta = pd.read_csv("../../data/ugc_meta/all_features.csv").set_index('FILENAME')
meta['category']=[str(meta.index[i]).split('_')[0] for i in range(meta.shape[0])]
meta
```

```python
df = meta.groupby('category').mean()[['SLEEQ_DMOS', 'BANDING_DMOS','SPATIAL_COMPLEXITY', 'TEMPORAL_COMPLEXITY', 'CHUNK_COMPLEXITY_VARIATION', 'COLOR_COMPLEXITY']].round(2)
df
```

```python
df['count'] = meta.groupby('category').count().iloc[:,0]
df
```
