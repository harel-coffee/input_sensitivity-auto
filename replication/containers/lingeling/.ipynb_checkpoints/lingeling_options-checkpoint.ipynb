### Goal : Generate the configurations for lingeling

##### Libraries

```python
import os, time 
import numpy as np
import pandas as pd
from scipy import stats
```

##### Runtime configs

```python
cols = ["--boost", "--carduse", "--decompose", "--gluescale", "--lkhd", "--memlim", 
       "--minimize", "--prbsimple", "--sweepirr", "--sweepred"]
dico = dict({"--boost" : ["0", "1"],
             "--carduse" : ["1", "2", "3"],
             "--decompose" : ["0","1"],
             "--gluescale" : ["1","2", "3", "4", "5"],
             "--lkhd" : ["-1", "0", "1", "2", "3", "4"],
             "--memlim" : ["-1", "1000"],
             "--minimize" : ["1","2"],
             "--prbsimple" : ["1","2","3"],
             "--sweepirr" : ["1","2","3"],
             "--sweepred" : ["1","2","3"]})
```

```python
df = pd.DataFrame({})

nb_conf = 150
nb_conf_final = 100

for d in dico.keys():
    df[d] = np.random.randint(0,len(dico[d]),nb_conf)

df = df.drop_duplicates()[0:nb_conf_final].reset_index().drop(["index"], axis=1)
df
```

```python
df.to_csv('runtime_options.csv')
```

```python
for j in range(nb_conf_final):
    with open("./scripts/"+str(j)+".sh","w") as f:
        f.write('#!/bin/bash\nnumb="'+str(j))
        f.write('"\nlogfilename="./logs/$numb.log"\ninputlocation="$1"\n')
        f.write("# execute the configuration \n")
        vals = df.iloc[j]
        cmd_line = "./lingeling/lingeling -T 10 "
        list_conf = ""
        for i in range(len(vals)):
            cmd_line+=cols[i]+"="+dico[cols[i]][vals[i]]+" "
            list_conf+=str(dico[cols[i]][vals[i]])+","*(i!=(len(vals)-1))
        cmd_line+= " $inputlocation > $logfilename\n"
        f.write(cmd_line)
        f.write("# analyze log to extract relevant information\n")
        f.write("""confs=`grep "confs/sec" $logfilename | sed 's/c//;s/confs\/sec//;s/conflicts//'`\n""")
        f.write("""reduc=`grep "reductions" $logfilename | sed 's/c//' | cut -d "r" -f 1`\n""")
        f.write('''csvLine="$numb,''')
        f.write(list_conf+'''"\n''')
        f.write('''csvLine="$csvLine,$confs,$reducs"\n''')
        f.write('''echo $csvLine''')
```
