# Runtime options

```python
runtime_options = ["--jitless","--experimental-wasm-modules","--experimental-vm-modules",
                   "--preserve-symlinks-main","--no-warnings","--node-memory-debug"]
```

```python
df = pd.DataFrame(
    np.random.randint(0,2,len(runtime_options)*200).reshape(200,len(runtime_options))).drop_duplicates()
df.columns = runtime_options
df.to_csv("runtime_options.csv")
```

#### Scripts nodejs

```python
cmd_line_node = "./node-15.14.0/out/Release/node"
csvLines = []
nb_config = 50
cols = df.columns

for j in range(nb_config):
    with open("./scripts/"+str(j)+".sh","w") as f:
        f.write('#!/bin/bash\n\nnumb="'+str(j))
        f.write('"\nlogfilename="./logs/$numb.log"\ninputconf="$1"\n\n')
        vals = df.iloc[j]
        cmd_line = "./node-15.14.0/out/Release/node "
        list_conf = ""
        for i in range(len(vals)):
            if vals[i]:
                cmd_line+=cols[i]+" "
            list_conf+=str(vals[i])+","
        cmd_line+= " ./node-15.14.0/benchmark/run.js all > $logfilename\n"
        f.write(cmd_line)
```

##### example of working script

#!/bin/bash

numb='1'
logfilename="./logs/$numb.log"
inputconf="$1"

./node-15.14.0/out/Release/node ./node-15.14.0/benchmark/$inputconf > $logfilename
# analyze log to extract operation per second
ops=`grep ":*" $logfilename | cut -f2 -d":"`

csvLine="$numb,"
csvLine="$csvLine$ops"
echo $csvLine
